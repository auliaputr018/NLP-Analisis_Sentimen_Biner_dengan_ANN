```python
from youtube_comment_downloader import YoutubeCommentDownloader
from youtube_comment_downloader import SORT_BY_POPULAR
import pandas as pd
```

```python
# link video YouTube
url = "https://www.youtube.com/watch?v=iAE-ltq8CFI"

# inisialisasi downloader
downloader = YoutubeCommentDownloader()

# mengambil komentar
comments = downloader.get_comments_from_url(
    url,
    sort_by=SORT_BY_POPULAR
)

# list untuk menyimpan data
data = []

# jumlah komentar yang ingin diambil
max_comments = 2000

# looping komentar
for i, comment in enumerate(comments):

    # menyimpan data komentar
    data.append({
        'username': comment['author'],
        'comment': comment['text'],
        'likes': comment['votes'],
        'time': comment['time']
    })

    # membatasi jumlah komentar
    if i + 1 >= max_comments:
        break

# membuat dataframe
df = pd.DataFrame(data)

# menyimpan ke CSV
df.to_csv('youtube_comments.csv', index=False, encoding='utf-8-sig')

print(f"\nBerhasil mengambil {len(df)} komentar")
print("File tersimpan: youtube_comments.csv")
```
