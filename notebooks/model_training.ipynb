### 1.4. Definisi Kelas Sentimen (Biner)

Agar lebih sederhana dan efektif, kita menggunakan **klasifikasi BINER**:

| Kelas | Definisi | Contoh Komentar |
|-------|----------|-----------------|
| **NEGATIF** | Komentar yang berisi kemarahan, kekecewaan, kritik pedas terhadap juri, MC, atau penyelenggara. Juga komentar yang menyebut kecurangan, ketidakadilan, arogansi kekuasaan. | *"Juri budeg! Dasar tidak kompeten. Malu-maluin acara sebesar ini."* |
| **POSITIF** | Komentar yang mendukung peserta SMAN 1 Pontianak, memuji keberanian mereka, memberikan semangat. Juga komentar netral seperti "up", "setuju", "ijin nyimak". | *"Semangat adik-adik SMAN 1 Pontianak! Kalian juara sejati di hati kami."* |

**Catatan:** Kelas POSITIF digabung dengan komentar NETRAL karena jumlah komentar netral sangat sedikit dan tidak signifikan. Fokus utama analisis adalah mengukur seberapa besar **kemarahan publik** (sentimen NEGATIF).

## 3.6 Preprocessing likes column

```python
# membersihkan kolom likes
def clean_likes(likes_str):
    """Fungsi untuk membersihkan format likes seperti '2 rb' menjadi 2000"""
    if pd.isna(likes_str):
        return 0
    if isinstance(likes_str, (int, float)):
        return likes_str
    
    likes_str = str(likes_str)
    likes_str = likes_str.replace(',', '').replace(' ', '')
    
    if 'rb' in likes_str:
        likes_str = likes_str.replace('rb', '')
        try:
            return float(likes_str) * 1000
        except:
            return 0
    else:
        try:
            return float(likes_str)
        except:
            return 0

df['likes_clean'] = df['likes'].apply(clean_likes)
df['likes_clean'] = df['likes_clean'].astype(int)

print("="*50)
print("KOLOM LIKES SETELAH DIBERSIHKAN")
print("="*50)
print(df[['likes', 'likes_clean']].head(10))
```

Insight:

Kolom likes berhasil dikonversi dari format teks menjadi numerik sehingga lebih mudah digunakan pada tahap analisis. Proses ini penting karena data hasil scraping masih mengandung format seperti “rb” yang tidak dapat langsung diproses secara numerik. Setelah dibersihkan, jumlah likes dapat dianalisis lebih akurat untuk melihat tingkat perhatian pengguna terhadap suatu komentar.

## 4.8 Extract time information

```python
def extract_days(time_str):
    """Function to extract number of days from format like 'X hari yang lalu'"""
    if pd.isna(time_str):
        return 0
    time_str = str(time_str)
    if 'hari' in time_str:
        numbers = re.findall(r'\d+', time_str)
        return int(numbers[0]) if numbers else 0
    elif 'jam' in time_str:
        return 0
    elif 'menit' in time_str:
        return 0
    else:
        return 0

df['days_ago'] = df['time'].apply(extract_days)

print("="*50)
print("TIME COLUMN AFTER EXTRACTION")
print("="*50)
print(df[['time', 'days_ago']].head(10))
```

### 4.3.2 Text Cleaning

```python
# Fungsi untuk membersihkan teks komentar
def clean_text(text):

    # Menghindari error jika data bukan string
    if not isinstance(text, str):
        return ""

    # Menghapus username seperti @namauser
    text = re.sub(r'@\w+', ' ', text)

    # Menghapus link
    text = re.sub(r'http\S+|www\S+', ' ', text)

    # Menghapus karakter enter, tab, dan newline
    text = re.sub(r'[\n\r\t]+', ' ', text)

    # Menghapus angka dan tanda baca
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)

    # Mengubah teks menjadi huruf kecil
    text = text.lower()

    # Menghapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    return text


# Melakukan proses cleaning pada kolom komentar
df_proc['comment_clean'] = df_proc['comment'].apply(clean_text)

# Menghapus komentar yang kosong setelah cleaning
df_proc = df_proc[df_proc['comment_clean'] != '']


# Menampilkan jumlah data setelah cleaning
print('=' * 50)
print('HASIL TEXT CLEANING')
print('=' * 50)

print(f'Jumlah komentar setelah cleaning : {len(df_proc)}')
print(f'Jumlah komentar yang dihapus     : {len(df) - len(df_proc)}')


# Menampilkan contoh hasil cleaning
for i in range(2):

    print(f'\n{"=" * 50}')
    print(f'CONTOH KOMENTAR {i+1}')

    print('\nSebelum Cleaning:')
    print(df_proc['comment'].iloc[i][:120])

    print('\nSesudah Cleaning:')
    print(df_proc['comment_clean'].iloc[i][:120])
```

## 4.4 Text Normalization

### 4.4.1 Convert Dictionary

```python
# Mengubah kamus normalisasi menjadi dictionary
slang_dict = dict(
    zip(
        kamus_df['tidak_baku'],
        kamus_df['kata_baku']
    )
)

print('=' * 50)
print('Normalization Dictionary')
print('=' * 50)

# Menampilkan jumlah pasangan kata
print(f'Jumlah pasangan kata dalam dictionary: {len(slang_dict)}')

# Menampilkan contoh isi dictionary
print('\nContoh pasangan kata:\n')

for slang, baku in list(slang_dict.items())[:10]:
    print(f"'{slang}' → '{baku}'")
```

Insight:

Dictionary normalisasi berhasil dibuat dengan total 4347 pasangan kata tidak baku dan kata baku. Hasil ini menunjukkan bahwa komentar pada media sosial memiliki cukup banyak variasi bahasa informal, singkatan, dan typo sehingga proses normalisasi diperlukan agar teks menjadi lebih konsisten sebelum digunakan pada analisis sentimen.

### 4.4.2 Text Normalization Process

```python
# Fungsi untuk menghapus huruf yang berulang berlebihan
def remove_repeated_chars(text):

    # Menghindari error jika data bukan string
    if not isinstance(text, str):
        return ""

    # Mengubah huruf berulang seperti 'bagussssss' menjadi 'bagus'
    return re.sub(r'(.)\1{2,}', r'\1', text)


# Fungsi untuk normalisasi teks
def normalize_text(text):

    # Menghindari error jika data bukan string
    if not isinstance(text, str):
        return ""

    # Menghapus huruf berulang
    text = remove_repeated_chars(text)

    # Memecah teks menjadi token kata
    words = text.split()

    # Mengganti kata tidak baku menggunakan dictionary
    normalized_words = [
        slang_dict.get(word, word)
        for word in words
    ]

    # Menggabungkan kembali kata menjadi kalimat
    return ' '.join(normalized_words)


# Contoh hasil normalisasi
sample_text = "gak mau bgt lucuuuuuu paraaaaaaaaaah"

print('=' * 50)
print('Normalization example')
print('=' * 50)

print(f'Original   : {sample_text}')
print(f'Normalized : {normalize_text(sample_text)}')


# Menerapkan normalisasi pada kolom komentar
df_proc['comment_normalized'] = df_proc['comment_clean'].apply(normalize_text)
```

Insight:

Proses normalisasi membantu mengurangi variasi kata tidak baku dan huruf berulang yang sering muncul pada komentar media sosial. Dengan normalisasi ini, teks menjadi lebih konsisten sehingga model lebih mudah memahami konteks komentar.

```python
# Mengambil stopwords bawaan Sastrawi
factory = StopWordRemoverFactory()
sastrawi_stopwords = set(factory.get_stop_words())


# Mempertahankan kata negasi agar makna sentimen tidak hilang
negation_words = {'tidak'}

# Menghapus kata negasi dari daftar stopwords
sastrawi_stopwords = sastrawi_stopwords - negation_words

# Stopwords tambahan untuk komentar YouTube
extra_stopwords = {
    'yg', 'nya', 'sih', 'dong', 'deh', 'nih', 'yuk', 'ayo', 'woi', 'coy',
    'jadi', 'kayak', 'kaya', 'iya', 'udah', 'udh', 'gitu', 'begitu', 'gini', 
    'begini', 'aja', 'ajah', 'doang', 'kan', 'kok', 'loh', 'nah', 'weh',
    'banget', 'bgt', 'bgtt', 'bngt', 'banged','sihh', 'sich', 'seh', 'dongg', 
    'dunk', 'deh', 'dah', 'doank', 'cuma', 'cuman', 'cm', 'cmn', 'emang', 'memang',
    'kayaknya', 'kyaknya', 'kayanya', 'sepertinya','wkwk', 'wkwkwk', 'haha', 'hehe', 
    'hihi', 'lol','orang', 'kalian', 'disini', 'd sini', 'anda', 'kita', 'kami',
    'kamu', 'mereka', 'siapa', 'apakah', 'dimana', 'kemana', 'dari mana',
    'begini', 'begitu', 'seperti', 'karena', 'sebab', 'sehingga',
    'maka', 'mungkin', 'banyak', 'sedikit', 'beberapa', 'semua',
    'seluruh', 'sendiri', 'sendirian', 'terus', 'lalu', 'kemudian',
    'setelah', 'sebelum', 'selama', 'ketika', 'saat', 'pada', 'dalam',
    'atas', 'bawah', 'antara', 'melalui', 'tanpa', 'oleh', 'kalau',
    'mau','sama','tuh','lah','bro','bang','kak','akun', 'apa'
}

# Menggabungkan seluruh stopwords
all_stopwords = sastrawi_stopwords.union(extra_stopwords)

print('=' * 50)
print('Stopwords configuration')
print('=' * 50)

print(f'Total stopwords: {len(all_stopwords)} kata')


# Fungsi untuk menghapus stopwords
def remove_stopwords(text):

    # Menghindari error jika data bukan string
    if not isinstance(text, str):
        return ""

    # Memecah teks menjadi token kata
    words = text.split()

    # Menghapus stopwords dari teks
    filtered_words = [
        word for word in words
        if word not in all_stopwords
    ]

    # Menggabungkan kembali token menjadi kalimat
    return ' '.join(filtered_words)


# Contoh hasil stopwords removal
sample_text = "saya sangat marah dengan juri yang tidak adil"

print('\nContoh stopwords removal:\n')

print(f'Original : {sample_text}')
print(f'Result   : {remove_stopwords(sample_text)}')


# Menerapkan stopwords removal pada dataset
df_proc['comment_final'] = df_proc['comment_normalized'].apply(remove_stopwords)

# Menghapus komentar kosong setelah stopwords removal
df_proc = df_proc[df_proc['comment_final'] != '']

print(f'\nJumlah data setelah stopwords removal: {len(df_proc)}')
```

Proses stopwords removal berhasil mengurangi kata-kata yang kurang informatif pada komentar sehingga teks menjadi lebih fokus pada kata yang memiliki makna penting. Kata negasi seperti “tidak” tetap dipertahankan karena dapat mempengaruhi makna sentimen dalam suatu komentar.

Penjelasan Hyperparameter

1. Embedding Dimension

    `EMBEDDING_DIM = 128`

    - Digunakan untuk menentukan ukuran vector representasi kata pada embedding layer.

    - Ukuran 128 dipilih karenacukup untuk merepresentasikan hubungan antar kata, namun masih ringan untuk proses training.

2. LSTM Units

    `LSTM_UNITS = 64`

    - Menentukan jumlah neuron pada layer LSTM.

    - Nilai 64 dipilih karena cukup untuk mempelajari pola sequence teks, tidak terlalu kecil, dan masih stabil untuk dataset berukuran menengah.

3. Dense Units

    `DENSE_UNITS = 32`

    - Digunakan untuk membantu model mempelajari pola non-linear setelah proses ekstraksi fitur dari LSTM.

4. Dropout Rate

    `DROPOUT_RATE = 0.2`

    - Dropout digunakan untuk membantu mengurangi overfitting dengan menonaktifkan sebagian neuron secara acak selama proses training.

5. Batch Size

    `BATCH_SIZE = 32`

    - Batch size menentukan jumlah data yang diproses pada setiap iterasi training.

    - Nilai 32 dipilih karena cukup stabil, umum digunakan pada deep learning, dan tidak terlalu membebani memori.

6. Epoch

    `EPOCHS = 30`

    - Jumlah epoch dibuat 30 agar performa model dapat terlihat lebih jelas selama proses training.

7. Learning Rate

    `LEARNING_RATE = 0.001`

    - Learning rate digunakan untuk mengatur kecepatan model dalam memperbarui bobot selama proses training.

    - Nilai 0.001 dipilih karena merupakan learning rate default Adam yang cukup stabil untuk kasus NLP.

##### 5.1.1.1.2 Sequential API Architecture

```python
# Membuat model Sequential API
sequential_model_exp1 = Sequential([

    # Input layer
    Input(shape=(MAX_SEQUENCE_LENGTH,)),

    # Embedding layer
    Embedding(
        input_dim=MAX_VOCAB_SIZE,
        output_dim=EMBEDDING_DIM
    ),

    # LSTM layer
    LSTM(
        LSTM_UNITS,
        dropout=DROPOUT_RATE,
        recurrent_dropout=DROPOUT_RATE
    ),

    # Dense layer
    Dense( DENSE_UNITS, activation='relu' ),

    # Output layer
    Dense( 1, activation='sigmoid' )
])

# Optimizer Adam
optimizer = Adam( learning_rate=LEARNING_RATE )

# Compile model
sequential_model_exp1.compile(
    optimizer=optimizer,
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        Precision(name='precision'),
        Recall(name='recall')
    ]
)

# Menampilkan arsitektur model
sequential_model_exp1.summary()
```

Penjelasan Arsitektur Sequential API

1. Embedding Layer

    Embedding layer digunakan untuk mengubah token numerik menjadi vector dense sehingga hubungan antar kata dapat dipelajari lebih baik oleh model.

    Layer ini menghasilkan output:

    `(None, 100, 128)`

    yang berarti:
    - sequence memiliki panjang 100 token,
    - dan setiap token direpresentasikan menjadi vector 128 dimensi.

2. LSTM Layer

    - LSTM digunakan karena mampu mempelajari pola sequence pada data teks komentar.

    - Layer ini membantu model memahami hubungan antar kata dalam suatu kalimat sehingga konteks komentar dapat dipelajari lebih baik dibanding Dense layer biasa.

3. Dense Layer

    Dense layer digunakan untuk membantu model mempelajari pola non-linear dari hasil ekstraksi fitur layer sebelumnya.

    Activation function ReLU dipilih karena:

    - proses komputasi lebih cepat,
    - mengurangi vanishing gradient,
    - dan umum digunakan pada hidden layer deep learning.

4. Output Layer

    Output layer menggunakan:

    - 1 neuron,
    - activation sigmoid,

    karena kasus yang diselesaikan merupakan klasifikasi biner:

    - negatif (0)
    - positif (1)

    Sigmoid menghasilkan output probabilitas antara 0–1.

### 6.1.1 Model Definition

#### 6.1.1.1 Hyperparameter Experiment 1

```python
# Hyperparameter 
EMBEDDING_DIM_IMP = 128
LSTM_UNITS_IMP = 128
DENSE_UNITS_IMP = 64
DROPOUT_RATE_IMP = 0.3
BATCH_SIZE_IMP = 32
EPOCHS_IMP = 30
LEARNING_RATE_IMP = 0.001
```

3. L2 Regularization

    digunakan untuk membatasi kompleksitas bobot model agar proses training lebih stabil dan model tidak terlalu mudah overfit.

#### 6.3.1.1 Hyperparameter Experiment 3

```python
VOCAB_SIZE = 10000
EMBEDDING_DIM = 64
MAX_LENGTH = 100
LSTM_UNITS = 64
DROPOUT_RATE = 0.3
RECURRENT_DROPOUT = 0.3
DENSE_DROPOUT = 0.5
LEARNING_RATE = 0.0005
BATCH_SIZE = 32
EPOCHS = 30
```

Penjelasan Hyperparameter
1. Embedding Dimension

    `EMBEDDING_DIM = 64`

    Digunakan untuk merepresentasikan kata ke dalam bentuk vector numerik agar hubungan antar kata dapat dipelajari model.

2. LSTM Units

    `LSTM_UNITS = 64`

    Digunakan untuk mempelajari pola sequence dan konteks pada komentar teks.

3. Dropout
    
    `DROPOUT_RATE = 0.3`

    Digunakan untuk membantu mengurangi overfitting dengan menonaktifkan sebagian neuron secara acak selama training.

4. Dense Dropout
    
    `DENSE_DROPOUT = 0.5`

    Dropout tambahan pada dense layer digunakan agar model tidak terlalu bergantung pada neuron tertentu.

5. Learning Rate
    `LEARNING_RATE = 0.0005`

    Learning rate dibuat lebih kecil agar proses update bobot menjadi lebih stabil dan model dapat belajar secara lebih optimal.

#### 6.3.1.2 Class Weight

```python
# Class weight
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)

class_weight_dict = {
    0: class_weights[0],
    1: class_weights[1]
}

print(class_weight_dict)
```

Insight:

Class weight tetap digunakan pada tahap improvement karena dataset memiliki distribusi kelas yang tidak seimbang. Penggunaan class weight membantu model memberikan perhatian lebih besar pada kelas minoritas sehingga prediksi sentimen menjadi lebih seimbang.

#### 6.3.1.3 Arsitektur Model Improvement Experiment 3

```python
# Input layer
inputs = Input(
    shape=(MAX_LENGTH,),
    name='input_layer'
)

# Embedding layer
x = Embedding(
    input_dim=VOCAB_SIZE,
    output_dim=EMBEDDING_DIM,
    name='embedding_layer'
)(inputs)

# Spatial Dropout
x = SpatialDropout1D(
    0.3,
    name='spatial_dropout'
)(x)

# Bidirectional LSTM
x = Bidirectional(
    LSTM(
        units=LSTM_UNITS,
        dropout=DROPOUT_RATE,
        recurrent_dropout=RECURRENT_DROPOUT,
        return_sequences=False,
        kernel_regularizer=l2(0.003)
    ),
    name='bidirectional_lstm'
)(x)

# Dropout layer
x = Dropout(
    rate=DENSE_DROPOUT,
    name='dropout_layer'
)(x)

# Output layer
outputs = Dense(
    units=1,
    activation='sigmoid',
    name='output_layer'
)(x)

# Build model
model_improved_exp3 = Model(
    inputs=inputs,
    outputs=outputs,
    name='Bidirectional_LSTM_GoodFit'
)

# Compile model
model_improved_exp3.compile(
    optimizer=Adam(
        learning_rate=LEARNING_RATE
    ),
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        Precision(name='precision'),
        Recall(name='recall')
    ]
)

print("="*50)
print("ARSITEKTUR MODEL IMPROVEMENT")
print("="*50)
model_improved_exp3.summary()
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=2,
    min_lr=1e-6,
    verbose=1
)
```

```python
train_acc_1 = history_imp_exp1.history['accuracy'][-1]
val_acc_1 = history_imp_exp1.history['val_accuracy'][-1]

train_loss_1 = history_imp_exp1.history['loss'][-1]
val_loss_1 = history_imp_exp1.history['val_loss'][-1]

print("="*50)
print("Fit Analysis Experiment 1")
print("="*50)

print(f"Train Accuracy      : {train_acc_1:.4f}")
print(f"Validation Accuracy : {val_acc_1:.4f}")

print(f"Train Loss          : {train_loss_1:.4f}")
print(f"Validation Loss     : {val_loss_1:.4f}")
```

Insight:

Pada experiment pertama, train accuracy mencapai sekitar 99% dengan train loss yang sangat rendah sebesar 0.0260. Namun, validation accuracy hanya berada di sekitar 67% dengan validation loss sebesar 0.5755. Gap yang cukup besar antara performa training dan validation menunjukkan bahwa model mengalami **overfitting**. Model mampu mempelajari data training dengan sangat baik, tetapi kemampuan generalisasi terhadap data validation masih rendah sehingga performa pada data baru belum optimal.
